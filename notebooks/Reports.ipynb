```python
from classes.stock_client import StockClient
from classes.quant_analytics import QuantAnalytics as qa
from modules.report import QuantReport
import modules.plots as plots
```

```python
client = StockClient(ttl_price=60, ttl_info=900)
quant = qa(client)
symbols = ["AAPL", "MSFT", "NVDA", "GOOGL", "JPM","INTC"]
```

```python
r = QuantReport(title="My Analysis", subtitle="Q4 2024", author="PM Team")
r.add_section("Performance")
r.add_text("Risk-adjusted returns for the selected basket.")
r.add_metrics({"CAGR": 0.23, "Sharpe": 1.42, "Max DD": -0.18}, ncols=3)
r.add_plot(plots.cumulative_returns(quant, symbols))
r.add_section("Risk")
r.add_plot(plots.correlation_heatmap(quant, symbols))
r.add_table(quant.metrics_df(symbols), caption="Key metrics")

r.save("reports/manual_report.html")
r.show()
```

```python
from modules.report import auto_report

portfolios={
   "tech":       ["AAPL", "MSFT", "NVDA", "GOOGL", "META"],
   "financials": ["JPM", "BAC", "GS", "MS", "WFC"],
   "healthcare": ["JNJ", "PFE", "ABBV", "MRK", "UNH"],

}

risk_free_rate: float = 0.01
period="2y" # 1y 2y 3y 5y

for name, symbols in portfolios.items():
    print (f'Generating report for {name}')

    auto_report(
    quant, client,
    symbols,
    title="Investment Ideas",
    subtitle=name,
    period=period,
    benchmark="SPY",
    risk_free_rate=risk_free_rate,
    include_backtest = True,
    include_frontier= True,
    include_fundamentals = True,
    include_montecarlo = True,
    mc_horizon = 252,
    mc_sims = 1000,
).save(f"reports/{name}.html")
```
