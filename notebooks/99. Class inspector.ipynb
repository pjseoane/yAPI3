```python
import inspect

def get_class_methods(clase):
    [m for m in dir(clase) if not m.startswith('_')]

    for name, method in inspect.getmembers(clase, predicate=inspect.ismethod):
        if not name.startswith('_'):
            print(name, inspect.signature(method))


def get_method_signature(cls, method_name):
    """Returns the signature of a specific method within a class."""
    method = getattr(cls, method_name)
    return inspect.signature(method)
```

```python
get_class_methods(quant)
```
