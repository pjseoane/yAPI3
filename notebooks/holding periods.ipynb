```python
from classes.stock_client import StockClient
from classes.quant_analytics import QuantAnalytics as qa
import modules.plots as plots
```

```python
client = StockClient(ttl_price=60, ttl_info=900)
quant = qa(client)
```

```python
# single stock — all entry dates
rets  = quant.rolling_returns("SPY", hold_years=5, period="20y")
print(f"Mean 5y return: {rets.mean():.1%}")
print(f"Win rate:       {(rets > 0).mean():.1%}")

# full stats dict
stats = quant.rolling_returns_stats("SPY", hold_years=5, period="20y")
print(f"Worst entry: {stats['worst_entry']}  ({stats['min_return']:+.1%})")
print(f"Best entry:  {stats['best_entry']}  ({stats['max_return']:+.1%})")
print(f"Prob of 2x:  {stats['prob_double']:.0%}")

# compare across stocks
df = quant.rolling_returns_bulk(["AAPL","MSFT","NVDA","SPY"], hold_years=5)
print(df.loc[["mean_return","win_rate","cagr_mean"]].T)
```

```python
# single stock
plots.rolling_returns(quant, "SPY", hold_years=5, period="20y").show()

# compare distributions — overlay AAPL and MSFT on the histogram
plots.rolling_returns(quant, "SPY", hold_years=5, period="20y",
    compare_symbols=["AAPL", "MSFT"]).show()

# different holding periods
for n in [1, 3, 5, 10]:
    plots.rolling_returns(quant, "SPY", hold_years=n, period="20y").show()
```
