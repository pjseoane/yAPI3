# Stock Client Class

## Imports

```python
from yfinance_api3.classes.stock_client import StockClient
```

```python
"""
TL settings (seconds)
    ----------------------
    ttl_price   : real-time / fast-changing data  (default  60 s)
    ttl_info    : company metadata                (default 600 s)
    ttl_history : historical OHLCV bars           (default 300 s)
    ttl_news    : news headlines                  (default 120 s)
"""
client = StockClient(ttl_price=60, ttl_info=900)
```

## Class methods

```python
# all public methods (no underscore)
[m for m in dir(client) if not m.startswith('_')]

# with signatures
import inspect

for name, method in inspect.getmembers(client, predicate=inspect.ismethod):
    if not name.startswith('_'):
        print(name, inspect.signature(method))
```
