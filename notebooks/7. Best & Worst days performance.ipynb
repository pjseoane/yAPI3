```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

detail = quant.best_worst_days_detail("SPY", period="20y", n=20)
best  = detail[detail["type"] == "best"]
worst = detail[detail["type"] == "worst"]

# test panel 2 in isolation
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=best["date"],
    y=best["return"],
    mode="markers",
    name="Best"
))
fig.add_trace(go.Scatter(
    x=worst["date"],
    y=worst["return"],
    mode="markers",
    name="Worst"
))
fig.update_xaxes(type="date")
fig.show()
```
