```python
# put this at the top of every notebook
import plotly.io as pio
pio.renderers.default = "notebook_connected"
```

```python
from yfinance_api3.modules import backtest, plots
from yfinance_api3.classes.stock_client import StockClient
from yfinance_api3.classes.quant_analytics import QuantAnalytics as qa
```

```python
client = StockClient(ttl_price=60, ttl_info=900)
quant = qa(client)
```

```python
symbols = ["AAPL", "MSFT", "NVDA", "GOOGL", "JPM"]
```

```python
# run a strategy
result = backtest.run(quant, symbols,
                      strategy=backtest.ma_crossover(fast=20, slow=50),
                      period="3y",
                      rebalance="monthly",
                      transaction_cost_bps=10,
                      benchmark="SPY")

print(result)                   # readable summary
print(result.metrics_df())      # vs benchmark table

plots.backtest(result).show()
```

```python
from yfinance_api3.modules import backtest, plots

mag7 = ["AAPL", "MSFT", "NVDA", "GOOGL", "META", "AMZN", "TSLA"]

result = backtest.run(
    quant, mag7,
    strategy=backtest.buy_and_hold(),   # equal weight, no rebalancing cost
    period="6mo",
    rebalance="monthly",                # rebalance back to equal weight monthly
    transaction_cost_bps=5,
    benchmark="SPY",
)

print(result)                           # CAGR, Sharpe, Max DD vs SPY
print(result.metrics_df())              # side-by-side table
plots.backtest(result).show()           # equity curve + drawdown panel
```
