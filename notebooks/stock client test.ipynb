```python
from classes.stock_client import StockClient
#from classes.quant_analytics import QuantAnalytics as qa
```

```python
"""
TL settings (seconds)
    ----------------------
    ttl_price   : real-time / fast-changing data  (default  60 s)
    ttl_info    : company metadata                (default 600 s)
    ttl_history : historical OHLCV bars           (default 300 s)
    ttl_news    : news headlines                  (default 120 s)
"""
client = StockClient(ttl_price=60, ttl_info=900)
```

```python
print(client.get_price("AAPL"))          # hits yfinance
print(client.get_price("AAPL"))
```

```python
print(client.get_history("TSLA", period="3mo", interval="1wk"))
```

```python
print(client.get_news("NVDA", limit=5))
```

```python
print(client.cache_stats())              # {"total_entries": 3, "alive": 3, ...}
```

```python
client.invalidate("AAPL")               # clear only AAPL entries
```

```python
client.invalidate()
```
