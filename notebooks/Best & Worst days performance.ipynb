```python
from yfinance_api3.classes.stock_client import StockClient
from yfinance_api3.classes.quant_analytics import QuantAnalytics as qa
import yfinance_api3.modules.plots as plots
```

```python
client = StockClient(ttl_price=60, ttl_info=900)
quant = qa(client)
```

```python
# main analysis table
df = quant.best_worst_days_impact("SPY", period="20y")
print(df[["scenario", "final_value", "total_return", "cagr"]].to_string())

# see exactly when the extreme days happened
detail = quant.best_worst_days_detail("SPY", period="20y", n=20)
print(detail)   # sorted best→worst, dates show they cluster around crises
```

```python
plots.best_worst_days(quant, "SPY", period="20y").show()

# custom scenarios
plots.best_worst_days(quant, "SPY",
                      miss_scenarios=[1, 5, 10, 20, 50],
                      initial_value=100_000).show()

# works for any stock/ETF
plots.best_worst_days(quant, "QQQ", period="20y").show()
```

## Concentration of Good / Bad days

```python
plots.extreme_days_concentration(quant, "SPY", period="20y").show()

# zoom in on fewer extreme days
plots.extreme_days_concentration(quant, "SPY", period="20y", n=20).show()

# shorter window for density (monthly instead of quarterly)
plots.extreme_days_concentration(quant, "SPY", n=50, window=21).show()
```
