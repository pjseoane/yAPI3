```python
df = quant.seasonality_holding_sharpe("SPY", period="20y", holding_years=[2, 5])
print(df.to_string())

# Best months to buy for 2-year hold
print("\nBest 2y entry months:")
print(df.sort_values("sharpe_2y", ascending=False))

# Best months for 5-year hold
print("\nBest 5y entry months:")
print(df.sort_values("sharpe_5y", ascending=False))

# Patient vs momentum months
print("\nPatient months (5y >> 2y):")
print(df[df["edge"] == "patient"])

print("\nMomentum months (2y >> 5y):")
print(df[df["edge"] == "momentum"])
```

```python
# ── 1. Holding-period Sharpe ──────────────────────────────────────────────
hs = quant.seasonality_holding_sharpe("SPY", period="20y", holding_years=[2, 5])
print("Holding Sharpe:")
print(hs.to_string())

# ── 2. Drawdown during hold ───────────────────────────────────────────────
dd = quant.seasonality_holding_drawdown("SPY", period="20y", holding_years=[2, 5])
print("\nDrawdown during hold:")
print(dd.to_string())

# ── 3. Decade analysis ────────────────────────────────────────────────────
dec = quant.seasonality_decade_analysis("SPY", period="30y", holding_years=2)
print("\nDecade analysis:")
print(dec.to_string())

# ── 4. Cross-asset comparison ─────────────────────────────────────────────
ca = quant.seasonality_cross_asset(
    ["SPY","QQQ","GLD","BTC-USD"],
    period="10y", holding_years=[2, 5]
)
print("\nCross-asset:")
print(ca.to_string())

# ── 5. Combined actionable score ──────────────────────────────────────────
score = quant.seasonality_combined_score("SPY", period="20y", holding_years=[2, 5])
print("\nCombined score (ranked):")
print(score.to_string())
# Top signal: score["signal"] == "strong buy"
```
