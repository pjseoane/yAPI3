```python
from classes.stock_client import StockClient
from classes.quant_analytics import QuantAnalytics as qa
import modules.plots as plots
import modules.portfolio as portfolio
import pandas as pd
```

```python
client = StockClient(ttl_price=60, ttl_info=900)
quant = qa(client)
```

```python
symbols = ["AAPL", "MSFT", "NVDA", "GOOGL", "JPM", "TSLA", "SPY"]

# single stock — full breakdown
result = quant.kelly("NVDA", period="2y", fractional=0.5)
print(result)

# basket ranked by edge
df = quant.kelly_bulk(symbols, period="2y", fractional=0.5)
print(df)

# Kelly-weighted portfolio (use as input to backtest or compare vs max_sharpe)
weights = quant.kelly_portfolio(symbols, period="2y", fractional=0.5)
print(weights)

# three-panel chart
plots.kelly(quant, symbols, period="2y", fractional=0.5).show()

# compare Kelly weights vs Max Sharpe
ms = portfolio.max_sharpe(quant, symbols, period="2y")
print(pd.DataFrame({
    "kelly":     quant.kelly_portfolio(symbols),
    "max_sharpe": pd.Series(ms.weights, index=symbols),
}))

```
