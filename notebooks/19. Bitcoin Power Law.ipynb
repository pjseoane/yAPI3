```python
from yfinance_api3.modules.powerlaw import PowerLaw
import yfinance_api3.modules.plots as plots

# fit the model
pl = PowerLaw(corridor_low=10, corridor_high=90)
pl.fit()

# summary
print(pl.summary())

# current position
pos = pl.current_position()
print(f"\nPhase: {pos['phase']}")
print(f"Position in corridor: {pos['corridor_pct']:.1%}")

# fair value projections
print(pl.fair_value("2025-12-31"))
print(pl.fair_value("2028-01-01"))

# cycle peaks/troughs
print(pl.cycle_analysis())

# plots
plots.powerlaw_chart(pl).show()
plots.powerlaw_residuals(pl).show()
plots.powerlaw_forecast(pl, years=4).show()
```
